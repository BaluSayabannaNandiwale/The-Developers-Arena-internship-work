```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/supermarket_sales.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
total_sales = df["Total"].sum()
print("Total Sales:", total_sales)
```

```python
product_sales = df.groupby("Product_Line")["Total"].sum()
product_sales.plot(kind="bar", title="Sales by Product Line")
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
daily_sales = df.groupby("Date")["Total"].sum()

daily_sales.plot(kind="line", title="Daily Sales Trend")
plt.show()
```

Electronics and Food categories generate the highest revenue.
Weekends show higher sales trends.
