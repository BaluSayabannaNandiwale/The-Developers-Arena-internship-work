```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/healthcare_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["Gender"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6),
    title="Patient Gender Distribution"
)
plt.ylabel("")
plt.show()
```

```python
df["Medical Condition"].value_counts().head(10).plot(
    kind="bar",
    figsize=(8,4),
    title="Top Medical Conditions"
)
plt.ylabel("Number of Patients")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
billing_avg = df.groupby("Medical Condition")["Billing Amount"].mean().sort_values(ascending=False)

billing_avg.plot(
    kind="bar",
    figsize=(9,4),
    title="Average Billing Amount by Medical Condition"
)
plt.ylabel("Average Billing Amount")
plt.show()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df["Length_of_Stay"])
plt.title("Hospital Stay Distribution")
plt.xlabel("Days")
plt.show()

```

```python
plt.figure(figsize=(6,4))
numeric_cols = ["Age", "Billing Amount", "Length of Stay"]
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Healthcare Variables")
plt.show()
```
